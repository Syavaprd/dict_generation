```python

result_dict = {}
from nltk.corpus import stopwords 
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer as lemmatizer
stop_words = set(stopwords.words('english')) 
import nltk

def clean_first(word):
    if word[0] == '_':
        return word[1:]
    return word

def construct_vocabulary(sentence_list):
    wn_lemmatizer = lemmatizer()
    from tqdm import tqdm
    for sentence in tqdm(sentence_list):
        words = nltk.word_tokenize(sentence)
        for word in words:
            #print(word)
            if len(word) < 4:
                continue
            if word in stop_words:
                continue
            word = wn_lemmatizer.lemmatize(word.lower())
            if len(wn.synsets(word)) > 0:
                result_dict[word] = set()
                for synset in wn.synsets(word):
                    result_dict[word].add(synset.lexname())
                    #print (f"{synset.lexname()} with def {synset.definition()}") 
                #print(result_dict[word])
```

```python
from tqdm import tqdm
with open('../input/preprocessed/train_idx.input.bpe', 'r') as file:
    indices = file.readlines()
with open('../input/right-part-supersense/supersenses_final.txt', 'r') as file:
    right_part = file.readlines()

supersense_dict = sup
        
def construct_result_dict(indices=indices, right_part=right_part, supersense_dict=supersense_dict):
    result_dict = {}
    for ind, line in tqdm(enumerate(indices)):
        line = line.split(' ')
        line[-1] = line[-1][:-1]
        line = [int(x) for x in line]
        line = tuple(line)
        sups = right_part[ind]
        sups = sups.split(' ')
        result_dict[line] = []
        for sup in sups[:-1]:
            result_dict[line].append(supersense_dict.index(sup))

    return result_dict
```
